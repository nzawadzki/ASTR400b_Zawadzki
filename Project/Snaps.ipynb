```python
#fuction that will calculate the oribts from the 800 snapshots
#input: galaxy name (galaxy), first snapshot (start), last snapshot (end), integer for counting intervals (n)
def OrbitCOM (galaxy, start, end, n):
    #file that will store the oribt
    fileout = "SnapData_" + "%s"%(galaxy) + ".txt"
    
    #initializing array that will store position and velocity variables of the COM for the galaxy at each snapshot 
    orbit = np.zeros(shape=(int(end/n)+1, 8))
    
    #variable needed for convergence of COM calculation in CenterOfMass (for M33 do 0.5,and 4), 1 and 2 for the others
    delta = 0.5
    VolDec = 4.0
    
    for i in range(start, end+n, n):
        #construct filename for the galaxy being read in
        #add a string of filenumber to the value 000
        ilbl = '000' + str(i)
        #remove all but last three digits
        ilbl = ilbl[-3:]
        filename = "%s_"%(galaxy) + ilbl + '.txt'
        
        #create center of mass object using disk particles
        COM = CenterOfMass(filename, 2)
        
        #store which snapshot
        orbit[i/n][0] = ilbl
        
        #store time in Gyr in first column of orbit array
        orbit[i/n][1] = COM.time.value/1000
        
        #store COM positions in orbit array
        orbit[i/n][2] = float(COM.COM_P(delta, VolDec)[0])
        orbit[i/n][3] = float(COM.COM_P(delta, VolDec)[1])
        orbit[i/n][4] = float(COM.COM_P(delta, VolDec)[2])
        
        #store COM velocity positions in orbit array
        orbit[i/n][5] = float(COM.COM_V()[0])
        orbit[i/n][6] = float(COM.COM_V()[1])
        orbit[i/n][7] = float(COM.COM_V()[2])
        
        print i
        
    #save orbit array to a file
    return np.savetxt(fileout, orbit, header='snap t x y z vx vy vz', comments ='#', fmt=['%.2f', '%.2f', '%.2f', '%.2f', '%.2f', '%.2f', '%.2f', '%.2f'])
```
