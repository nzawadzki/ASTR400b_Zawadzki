```python
#necessary modules
import numpy as np
import matplotlib.pyplot as plt
get_ipython().magic('matplotlib inline')
```

```python
#import needed files
from ReadFile import *
from CenterOfMass import *
```

```python
#read in COM data files for MW, M31, and M33
fMW = np.genfromtxt('Orbit_MW.txt', dtype=None, names=True)
fM31 = np.genfromtxt('Orbit_M31.txt', dtype=None, names=True)
fM33 = np.genfromtxt('Orbit_M33.txt', dtype=None, names=True)
```

```python
#plot magnitude of separation for MW and M31 throughout the simulation
#subtract the x, y, and z positions of the COM of the galaxies
x = abs(fMW['x'] - fM31['x'])
y = abs(fMW['y'] - fM31['y'])
z = abs(fMW['z'] - fM31['z'])
r = np.sqrt(x**2 + y**2 + z**2)
#define time
time = fMW['t']
#test print statement
#print len(time), len(r)
#plot the separation vs time
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
plt.plot(time,r, linewidth=5)
plt.xlabel('Time (Gyr)', fontsize=22)
plt.ylabel('Separation (kpc)', fontsize=22)
plt.title('MW/M31 Separation', fontsize=22)
```

```python
#plot magnitude of separation for M31 and M33 throughout the simulation
#subtract the x, y, and z positions of the COM of the galaxies
x = abs(fM31['x'] - fM33['x'])
y = abs(fM31['y'] - fM33['y'])
z = abs(fM31['z'] - fM33['z'])
r = np.sqrt(x**2 + y**2 + z**2)
#define time
time = fM31['t']
#test print statement
#print len(time), len(r)
#plot the separation vs time
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
plt.plot(time,r, linewidth=5)
plt.xlabel('Time (Gyr)', fontsize=22)
plt.ylabel('Separation (kpc)', fontsize=22)
plt.title('M31/M33 Separation', fontsize=22)


#solving bonus question
#indexing time after 6 Gyr
#ind = np.where(time>6)
#using index for separation array
#rnew = r[ind]
#timenew = time[ind]
#looping over new r array 
#for i in range(len(rnew)):
    #if the middle separation is the greatest, this is the apocenter, and print the time and position
    #try: 
        #if rnew[i-2]<rnew[i-1] and rnew[i-1]>rnew[i]:
            #print 'position:', rnew[i-1], 'time:', timenew[i-1]
        #else:
            #a=0 #else do nothing
    #except: 
        #a=0
#test to make sure the loop worked
#plt.plot([6.07,6.07], [r[0], r[len(r)-1]], ls='--')
#plt.plot([7.5,7.5], [r[0], r[len(r)-1]], ls='--')
#plt.plot([8.93,8.93], [r[0], r[len(r)-1]], ls='--')
#plt.plot([10.07,10.07], [r[0], r[len(r)-1]], ls='--')
#plt.plot([11.07,11.07], [r[0], r[len(r)-1]], ls='--')
#test confirms positions of apocenters
```

```python
#plot magnitude of the relative velocities of MW and M31 throughout the simulation
#subtract the vx, vy, and vz positions of the COM of the galaxies
vx = abs(fMW['vx'] - fM31['vx'])
vy = abs(fMW['vy'] - fM31['vy'])
vz = abs(fMW['vz'] - fM31['vz'])
v = np.sqrt(vx**2 + vy**2 + vz**2)
#define time
time = fMW['t']
#test print statement
#print len(time), len(r)
#plot the velocity vs time
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
plt.plot(time,v, linewidth=5)
plt.xlabel('Time (Gyr)', fontsize=22)
plt.ylabel('Velocity (km/s)', fontsize=22)
plt.title('MW/M31 Relative Velocity', fontsize=22)
```

```python
#plot magnitude of the relative velocities of M31 and M33 throughout the simulation
#subtract the vx, vy, and vz positions of the COM of the galaxies
vx = abs(fM31['vx'] - fM33['vx'])
vy = abs(fM31['vy'] - fM33['vy'])
vz = abs(fM31['vz'] - fM33['vz'])
v = np.sqrt(vx**2 + vy**2 + vz**2)
#define time
time = fM31['t']
#test print statement
#print len(time), len(r)
#plot the separation vs time
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
plt.plot(time,v, linewidth=5)
plt.xlabel('Time (Gyr)', fontsize=22)
plt.ylabel('Velocity (km/s)', fontsize=22)
plt.title('M31/M33 Relative Velocities', fontsize=22)
```
